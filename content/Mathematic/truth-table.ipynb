## Conditional

| $p$ | $q$ | $p\to q$ |
|:---:|:---:|:---:|
| 1 | 1 | 1 |
| 1 | 0 | 0 |
| 0 | 1 | 1 |
| 0 | 0 | 1 |

!!!

$p$ = "you win"  
$q$ = "I give you money"  
$p\to q$ = "if you win then I give you money."  
What would make this statement false? i.e. $\neg (p \to q)$?  
Well, it is false only when, "if you win *and then* I **don't** give you money"  
$\therefore$ "If you lose and I give you money", the statement is still ok! I didn't break any promise!

!!!

a conditional is equivalent to its contrapositive.  
$\neg q \to \neg p$ = "If I didn't give you money then you must not have won"  
now look back to truth table of $p \to q$ and notice that if p then not q is false. Block that line and read the table, which states that the if you didn't money then you must not have won base from the condition. So ($\neg q \to \neg p$) is a true statement.

Can you do the same to prove why a conditional is not equal to its inverse or converse? i.e. prove the statement would be false.

## Biconditional

iff

| $p$ | $q$ | $p\leftrightarrow q$ |
|:---:|:---:|:---:|
| 1 | 1 | 1 |
| 1 | 0 | 0 |
| 0 | 1 | 0 |
| 0 | 0 | 1 |
$$
\begin{align*}
p\leftrightarrow q &= (p \to q ) \land (q \to p) \\
&= (p \to q) \land (\neg p \to \neg q ) \text{, constrapositive is true}
\end {align*}
$$
